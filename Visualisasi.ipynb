```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Visualisasi Jumlah actual train data mouth

```python
data_path = 'dataset_self_f1eye_aug.csv'
data = pd.read_csv(data_path)
print(f"Total frame folder for training: {len(data)}")
```

```python
class_distribution = data['class_name'].value_counts()
class_distribution
```

```python
colors = ['blue', 'orange', 'green', 'purple']

# Create a bar chart to visualize the class distribution
plt.figure(figsize=(10, 6))
bars = plt.bar(class_distribution.index, class_distribution.values, color=colors)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f'{yval}', ha='center', va='bottom')
    
plt.xlabel('Class Name')
plt.ylabel('Total Frame')
plt.title('Distribution of Augmented Mouth Dataset')
plt.xticks(rotation=360)  # Rotate class names for better visibility
plt.show()
```
